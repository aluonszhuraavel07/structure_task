```python
import random
rows = int(input("  : "))
cols = int(input("  : "))
matrix = []
print(" :")
for i in range(rows):
    row = [random.randint(-50, 50) for _ in range(cols)]
    matrix.append(row)
    print(row)
positive_count = 0
for row in matrix:
    for element in row:
        if element > 0:
            positive_count += 1
print("\n  :", positive_count)
sorted_matrix = [list(row) for row in matrix]
for j in range(cols):
    column = [sorted_matrix[i][j] for i in range(rows)]
    if (j + 1) % 2 == 0:
        column.sort(reverse=True)
    else:
        column.sort()
    for i in range(rows):
        sorted_matrix[i][j] = column[i]
print("\n   :")
for row in sorted_matrix:
    print(row)
```
